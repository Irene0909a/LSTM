```python
raw = pd.read_csv('/content/gdrive/MyDrive/科技部計畫/variables/各區間資料/MonthPredictMonth.csv',parse_dates=["Date"],index_col=[0])

#read the data
stock = raw 
amount = len(stock) 

stock = stock['2017-01-01':]
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn import datasets
import math 
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

y = stock["Close"]
x = stock

split =0.8
X_train,Y_train = x[0:int(split*len(x))] , y[0:int(split*len(x))]
X_test ,Y_test = x[int(split*len(x)):] , y[int(split*len(x)):]
##################################################################
lasso = Lasso(max_iter=10e5) # 默認alpha =1 
lasso.fit(X_train,Y_train)

train_score=lasso.score(X_train,Y_train)
test_score=lasso.score(X_test,Y_test)
lasso.feature_names_in_
coeff_used = np.sum(lasso.coef_!=0)
print("training score:", train_score )
print ("test score: ", test_score)
print ("number of features used: ", coeff_used)
keep_cols = [feature for feature, weight in zip (x.columns, lasso.coef_) if weight != 0]
keep_cols


```

```python
#找引述最佳值
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.linear_model import LassoCV
lassocv = LassoCV(cv=5, random_state=0, max_iter=10000)
lassocv.fit(X_train, Y_train)
al = lassocv.alpha_
print(al)

lasso = Lasso(alpha=al)
lasso.fit(X_train, Y_train)
print(lasso.score(X_test, Y_test)*100)
```
